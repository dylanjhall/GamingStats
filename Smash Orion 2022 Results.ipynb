```python
firsthalf2022["characters"] = firsthalf2022.index
secondhalf2022["characters"] = secondhalf2022.index
```

# Now lets see what the data looks like!

### visualization of Character Points vs Number of players
#### each colored shape represents a character

```python
plt.figure(figsize=(12,12),dpi=300)
sns.set_style("darkgrid")
sns.scatterplot(x="count",y="sum",data=firsthalf2022, palette="dark", style="characters",hue="characters")
plt.title("Character Resutls For The First Half Of 2022")
plt.xlabel("Number of Top Players")
plt.ylabel("Total Character Results")
plt.legend().remove()
```

```python
plt.figure(figsize=(12,12),dpi=300)
sns.set_style("darkgrid")
sns.scatterplot(x="count",y="sum",data=secondhalf2022, palette="dark", style="characters",hue="characters")
plt.title("Character Resutls For The Second Half Of 2022")
plt.xlabel("Number of Top Players")
plt.ylabel("Total Character Results")
plt.legend().remove()
```

## With these scatter plots we can see a strong trend of the more players a character has the more likely to get larger amount of points.
